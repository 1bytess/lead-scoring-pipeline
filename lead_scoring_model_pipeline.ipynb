# Lead Scoring Case Study: A Machine Learning Pipeline

**Objective**: Build a logistic regression model to assign a lead score to each potential customer. This score will help the sales team prioritize leads who are most likely to convert, optimizing their efforts and increasing sales efficiency.

### Step 1: Setup and Data Loading

First, we'll install and import the necessary libraries. We will then load the data from our PostgreSQL database using the secure environment variable method.

```python
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv
from pathlib import Path
from sqlalchemy import create_engine

# Visualization Libaries
import matplotlib.pyplot as plt
import seaborn as sns

# Scikit-learn for modeling
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve

#Set some display options for pandas
pd.set_option('display.max_columns', None)
```

```python
# --- Load Environment Variables and Connect to Database ---
env_path = Path('.') / '.env'
load_dotenv(dotenv_path=env_path)

# --- Database Connection Details (Loaded from .env file) ---
db_user = os.getenv('DB_USER')
db_password = os.getenv('DB_PASSWORD')
db_host = os.getenv('DB_HOST')
db_port = os.getenv('DB_PORT')
db_name = os.getenv('DB_NAME')

leads_df = pd.DataFrame()

if not all([db_user, db_password, db_host, db_name]):
    print("Error: Database credentials not found. Make sure your .env file is correct and saved.")
else:
    try:
        db_url = f"postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}"
        engine = create_engine(db_url)
        query = "SELECT * FROM leads;"
        leads_df = pd.read_sql(query, engine)
        print("Successfully loaded data into a pandas DataFrame!")
        print(f"\nDataFrame shape: {leads_df.shape}")

    except Exception as e:
        print(f"An error occurred: {e}")
          
```

### Step 2: Exploratory Data Analysis (EDA) and Cleaning

Now we'll explore the data to understand its structure, identify missing values, and clean it up for modeling.

```python
# Display the first 10 rows
leads_df.head()
```

```python
# Get a concise summary of the dataframe
leads_df.info()
```

```python
# Check for missing values
missing_values = leads_df.isnull().sum().sort_values(ascending=False)
missing_percentage = (leads_df.isnull().sum() / len(leads_df) * 100).sort_values(ascending=False)
missing_data_summary = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage}).round(2)
print("Missing Value Summary")
missing_data_summary[missing_data_summary['Missing Values'] > 0]
```

---
Based on the `missing_data_summary`, some columns have a very high percentage of missing values.

Let's drop columns with >40% nulls that are less likely to be predictive or are from external sources.

```python
cols_to_drop = [
    'Asymmetrique Activity Index', 'Asymmetrique Profile Index',
    'Asymmetrique Activity Score', 'Asymmetrique Profile Score',
    'How did you hear about X Education',
    'What matters most to you in choosing a course',
    'Lead Profile'
]
```

```python
leads_df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
print(f"Dropped columns with high nulls or low utility: {cols_to_drop}")
```

----

For important categorical features with missing values, we'll impute them.

Using a placeholder `Not Specified` can be a good strategy as missing data can be informative.

```python
for col in ['Specialization', 'What is your current occupation', 'Tags', 'Lead Quality']:
    if col in leads_df.columns:
        leads_df[col].fillna('Not Specified', inplace=True)
        print(f"Imputed '{col}' with 'Not Specified'")
```

---
For `City`, using the most frequent value (mode) is a safe bet.

```python
if 'City' in leads_df.columns:
    city_mode = leads_df['City'].mode()[0]
    leads_df['City'].fillna(city_mode, inplace=True)
    print(f"Imputed 'City' with mode: '{city_mode}'")
```

---
Impute numerical columns with the median as it's robust to outliers.

```python
for col in ['TotalVisits', 'Page Views Per Visit']:
    if col in leads_df.columns:
        median_val = leads_df[col].median()
        leads_df[col].fillna(median_val, inplace=True)
        print(f"Imputed '{col}' with median: {median_val}")
```

---
For columns with very few missing values, it's safe to drop the rows.

```python
leads_df.dropna(inplace=True)

print(f"Data cleaning complete. New shape: {leads_df.shape}")
print("Verifying null values:")
print(leads_df.isnull().sum())
```

---
Let's check the conversion rate

```python
conversion_rate = (leads_df['Converted'].sum() / len(leads_df)) * 100
print(f"Overall Conversion Rate: {conversion_rate:.2f}%")
```

### Step 3 : Feature Engineering & Selection

We will now prepare our features for the model. This involves converting categorical variables into a numerical format and selecting the most relevant features.

---
Define the target variable (y) and features (X)

```python
y = leads_df['Converted']
X = leads_df.drop(['Converted', 'Prospect ID', 'Lead Number'], axis=1)
```

```python
# Identify categorical and numerical columns
categorical_cols = X.select_dtypes(include=['object']).columns
numerical_cols = X.select_dtypes(include=np.number).columns

print("Categorical Columns:", list(categorical_cols))
print("Numerical Columns:", list(numerical_cols))
```

---
Let's look at the number of unique values in each categorical column.

This helps us understand the dimensionality after one-hot encoding

```python
print("Unique values in each categorical column:")
for col in categorical_cols:
    print(f"- {col}: {X[col].nunique()} unique values")
```

---
Some columns may have only one value and provide no information. Let's find and drop them

```python
cols_to_drop_low_variance = []
for col in X.columns:
    if X[col].nunique() <= 1:
        cols_to_drop_low_variance.append(col)

print(cols_to_drop_low_variance)
```

```python
if cols_to_drop_low_variance:
    X.drop(columns=cols_to_drop_low_variance, inplace=True)

    # We need to redefine categorical_cols after dropping some
    categorical_cols = X.select_dtypes(include=['object']).columns
    print("Updated Categorical Columns List:", list(categorical_cols))
else:
    print("No columns with only one unique value found.")
```

--- 
Preprocessing Pipeline

We'll use OneHotEncoder for categorical variables and StandardScaler for numerical ones.

```python
# Create preprocessing steps
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore') # Ignore new categories during prediction

# Create a preprocessor object using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers = [
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ],
    remainder = 'passthrough'
)
```

### Step 4: Model Building

It's time to build our model! We will split the data into training and testing sets and then train a Logistic Regression model.

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Training set shape: {X_train.shape}")
print(f"Testing set shape: {X_test.shape}")
```

---
Next, Apply the preprocessing pipeline to our data

We fit AND transform the training data, but only TRANSFORM the test data

```python
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

print("Data preprocessing complete.")
```

---
Now we initialize and train the Logistic Regression model

```python
log_reg = LogisticRegression(solver='liblinear', random_state=42)
log_reg.fit(X_train_processed, y_train)

print("Logistic Regression model train successfully.")
```

### Step 5: Model Evaluation

Let's see how well our model performs on the unseen test data.

---
Let's try make predictions on the test set

```python
y_pred = log_reg.predict(X_test_processed)

# --- Performance Metrics ---
print("Model Evaluation on Test Data:")
print("-" * 30)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4}")
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confussion Matrix")
plt.show()
```

```python
# --- ROC Curve and AUC Score ---
# Get predicted probabilities for the positive class (1)
y_pred_proba = log_reg.predict_proba(X_test_processed)[:, 1]\

auc_score = roc_auc_score(y_test, y_pred_proba)
print(f"ROC AUC Score: {auc_score:.4f}")
```

```python
# Plot the ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"Logistic Regression (AUC = {auc_score:.2f})")
plt.plot([0,1], [0,1], "k--")
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

## Step 6: Lead Scoring

Finally, we can use our trained model to assign a 'lead score' (a probability of conversion) to each lead.

```python
# Use the entire dataset for final scoring demonstration
X_full_processed = preprocessor.transform(X)
lead_scores = log_reg.predict_proba(X_full_processed)[:, 1]

# Add the lead score to the original DataFrame (matching by index)
X['lead_score'] = (lead_scores * 100).round(2)
X['converted_prediction'] = log_reg.predict(X_full_processed)

print("Lead scores and predictions added. Displaying top leads with highest scores:")
X_sorted_by_score = X.sort_values(by='lead_score', ascending=False)
X_sorted_by_score.head(10)
```
